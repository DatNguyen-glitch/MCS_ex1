```python
def convert_si_value(val):
    """Chuyn i gi tr c hu t SI thnh s th."""
    try:
        # Loi b khong trng v kim tra gi tr
        val = val.strip()
        if not val:  # Nu gi tr rng
            return None
        
        # nh ngha cc hu t SI v h s tng ng
        si_prefixes = {
            'p': 1e-12,  # pico
            'n': 1e-9,   # nano
            'u': 1e-6,   # micro
            'm': 1e-3,   # milli
            '':  1,      # khng hu t
            'k': 1e3,    # kilo
            'M': 1e6,    # mega
            'G': 1e9     # giga
        }
        
        # Tch s v hu t
        for prefix, factor in si_prefixes.items():
            if val.endswith(prefix):
                num_str = val[:-len(prefix)] if prefix else val
                return float(num_str) * factor
        return float(val)  # Nu khng c hu t, chuyn trc tip
    except ValueError:
        return val  # Tr v nguyn bn nu khng chuyn i c

def process_lis_file(input_file, output_file=None, start_line=50, end_line=7074):
    """c file .lis, x l t dng start_line n end_line, v chuyn i hu t SI."""
    with open(input_file, 'r') as f:
        lines = f.readlines()
    
    # Gii hn phm vi t start_line n end_line (chuyn v index 0-based)
    start_idx = start_line - 1
    end_idx = min(end_line, len(lines))  # m bo khng vt qu  di file
    data_lines = lines[start_idx:end_idx]
    
    # X l tng dng
    processed_lines = []
    for line in data_lines:
        # Tch cc ct bng khong trng
        columns = line.split()
        if columns:  # Ch x l dng khng rng
            # Chuyn i tng gi tr trong dng
            converted_columns = [str(convert_si_value(col)) if convert_si_value(col) is not None else col for col in columns]
            processed_lines.append(' '.join(converted_columns))
    
    # In ra mn hnh
    for line in processed_lines:
        print(line)
    
    # Nu c output_file, ghi vo file
    if output_file:
        with open(output_file, 'w') as f:
            f.write('\n'.join(processed_lines))

# S dng script
input_file = 'ex1_1.lis'  # Thay bng tn file .lis ca bn
output_file = 'converted_output.txt'  # File u ra (ty chn)

process_lis_file(input_file, output_file, start_line=50, end_line=7073) # Thay i start_line v end_line nu cn
```
